```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
A = 0.8
F = 5
T = np.arange(0,10,0.01)
Phi = np.pi/4

amp = A*np.sin(2*np.pi*F*T + Phi)
```

```python
plt.plot(T,amp)
plt.axis([0,1,-1,1])
plt.title('sine')
plt.show()
```

```python
tim = np.arange(-5,5,1)
EX = np.exp(tim)
```

```python
plt.stem(tim, EX, '-.')

plt.show()
```

```python
tim = np.arange(-5,5,1)
UR = []
for i in range(len(tim)):
    if tim[i] >= 0:
        UR.append(tim[i])
    else:
        UR.append(0)
```

```python
plt.stem(tim, UR, '-.')

plt.show()
```
